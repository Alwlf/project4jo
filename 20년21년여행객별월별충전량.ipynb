```python
total2020 = pd.read_csv('monthly_charging_by_station2020.csv')
total2021 = pd.read_csv('monthly_charging_by_station2021.csv')
```

```python
total2020 = total2020.drop('Unnamed: 0', axis=1)
total2021 = total2021.drop('Unnamed: 0', axis=1)
```

```python
total2020
```

```python
merged_charge = pd.concat([total2020, total2021], ignore_index=True)
merged_charge
```

```python
tour = pd.read_csv('20,21년관광객정리.csv')
```

```python
tour = tour.drop('Unnamed: 0', axis=1)
tour
```

```python
seogwipo = merged_charge[merged_charge['구'] == '서귀포시']
seogwipo['날짜'] = pd.to_datetime(merged_charge['month']).dt.month
result_seo = seogwipo.groupby('month')[['급속','완속']].sum()
result_seo
```

```python
seogwipo = merged_charge[merged_charge['구'] == '제주시']
seogwipo['날짜'] = pd.to_datetime(merged_charge['month']).dt.month
result_jeju = seogwipo.groupby('month')[['급속','완속']].sum()
result_jeju
```

```python
df = pd.concat([result_seo, result_jeju], axis=1, keys=['서귀포시', '제주시'])
df = pd.concat([df,tour], axis=1)
df.head()
```
