```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data1=pd.read_csv("data\한국전력공사_전기차충전소충전량(202001_202006).csv", encoding="cp949")
data2=pd.read_csv("data\한국전력공사_전기차충전소충전량(202007_202012).csv", encoding="cp949")
```

```python
data = pd.concat([data1, data2], ignore_index=True)
data
```

```python
data[['시','구','도로명']]=data["주소"].str.split(' ', n=2, expand=True)
data
```

```python
s=data[data["시"]=="서울특별시"]
s.reset_index(drop=True, inplace=True)
s
```

```python
s[['연','월','일']]=s["충전종료일자"].str.split('-', expand=True)
s
```

```python
rkdska=s[s["구"]=="강남구"]
rkdska.reset_index(drop=True, inplace=True)
rkdehd=s[s["구"]=="강동구"]
rkdehd.reset_index(drop=True, inplace=True)
rkdtj=s[s["구"]=="강서구"]
rkdtj.reset_index(drop=True, inplace=True)
rhksdkr=s[s["구"]=="관악구"]
rhksdkr.reset_index(drop=True, inplace=True)
rhkdwls=s[s["구"]=="광진구"]
rhkdwls.reset_index(drop=True, inplace=True)
rnfh=s[s["구"]=="구로구"]
rnfh.reset_index(drop=True, inplace=True)
rmacjs=s[s["구"]=="금천구"]
rmacjs.reset_index(drop=True, inplace=True)
shdnjs=s[s["구"]=="노원구"]
shdnjs.reset_index(drop=True, inplace=True)
ehqhd=s[s["구"]=="도봉구"]
ehqhd.reset_index(drop=True, inplace=True)
ehdeoans=s[s["구"]=="동대문구"]
ehdeoans.reset_index(drop=True, inplace=True)
ehdwkr=s[s["구"]=="동작구"]
ehdwkr.reset_index(drop=True, inplace=True)
akvh=s[s["구"]=="마포구"]
akvh.reset_index(drop=True, inplace=True)
tjeoans=s[s["구"]=="서대문구"]
tjeoans.reset_index(drop=True, inplace=True)
tjch=s[s["구"]=="서초구"]
tjch.reset_index(drop=True, inplace=True)
tjdehd=s[s["구"]=="성동구"]
tjdehd.reset_index(drop=True, inplace=True)
tjdqnr=s[s["구"]=="성북구"]
tjdqnr.reset_index(drop=True, inplace=True)
thdvk=s[s["구"]=="송파구"]
thdvk.reset_index(drop=True, inplace=True)
didcjs=s[s["구"]=="양천구"]
didcjs.reset_index(drop=True, inplace=True)
dudemdvh=s[s["구"]=="영등포구"]
dudemdvh.reset_index(drop=True, inplace=True)
dydtks=s[s["구"]=="용산구"]
dydtks.reset_index(drop=True, inplace=True)
dmsvud=s[s["구"]=="은평구"]
dmsvud.reset_index(drop=True, inplace=True)
wnd=s[s["구"]=="중구"]
wnd.reset_index(drop=True, inplace=True)
wndfkd=s[s["구"]=="중랑구"]
wndfkd.reset_index(drop=True, inplace=True)
whdfh=s[s["구"]=="종로구"]
whdfh.reset_index(drop=True, inplace=True)
```

```python
rkdska=rkdska.groupby('월')['충전량'].sum()
rkdehd=rkdehd.groupby('월')['충전량'].sum()
rkdtj=rkdtj.groupby('월')['충전량'].sum()
rhksdkr=rhksdkr.groupby('월')['충전량'].sum()
rhkdwls=rhkdwls.groupby('월')['충전량'].sum()
rnfh=rnfh.groupby('월')['충전량'].sum()
rmacjs=rmacjs.groupby('월')['충전량'].sum()
shdnjs=shdnjs.groupby('월')['충전량'].sum()
ehqhd=ehqhd.groupby('월')['충전량'].sum()
ehdeoans=ehdeoans.groupby('월')['충전량'].sum()
ehdwkr=ehdwkr.groupby('월')['충전량'].sum()
akvh=akvh.groupby('월')['충전량'].sum()
tjeoans=tjeoans.groupby('월')['충전량'].sum()
tjch=tjch.groupby('월')['충전량'].sum()
tjdehd=tjdehd.groupby('월')['충전량'].sum()
tjdqnr=tjdqnr.groupby('월')['충전량'].sum()
thdvk=thdvk.groupby('월')['충전량'].sum()
didcjs=didcjs.groupby('월')['충전량'].sum()
dudemdvh=dudemdvh.groupby('월')['충전량'].sum()
dydtks=dydtks.groupby('월')['충전량'].sum()
dmsvud=dmsvud.groupby('월')['충전량'].sum()
wnd=wnd.groupby('월')['충전량'].sum()
wndfkd=wndfkd.groupby('월')['충전량'].sum()
whdfh=whdfh.groupby('월')['충전량'].sum()
```
