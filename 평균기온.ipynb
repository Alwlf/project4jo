```python
import pandas as pd
import folium 
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc # 폰트 세팅을 위한 모듈 추가
font_path = "C:/Windows/Fonts/malgun.ttf" # 사용할 폰트명 경로 삽입
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)

import warnings
warnings.filterwarnings("ignore")

 
```

```python
file_path = "./2020_2023_월간평균기온.xlsx"
temp = pd.read_excel(file_path)
temp
```

```python
temp = temp.dropna()
```

```python
# 연도별 데이터 추출
temp_2020 = temp[temp['년월'].dt.year == 2020]
temp_2021 = temp[temp['년월'].dt.year == 2021]
temp_2022 = temp[temp['년월'].dt.year == 2022]
temp_2023 = temp[temp['년월'].dt.year == 2023]
```

```python
temp_2020['월'] = temp_2020['년월'].dt.strftime('%m')
temp_2021['월'] = temp_2021['년월'].dt.strftime('%m')
temp_2022['월'] = temp_2022['년월'].dt.strftime('%m')
temp_2023['월'] = temp_2023['년월'].dt.strftime('%m')
```

```python
temp_2020
```

```python
fig = plt.figure(figsize=(14,7))
fig.suptitle("연간 평균기온")

ax1 =fig.add_subplot(2,2,1)
ax2 =fig.add_subplot(2,2,2)
ax3 =fig.add_subplot(2,2,3)
ax4 =fig.add_subplot(2,2,4)

ax1.plot(temp_2020['월'],temp_2020['평균기온(℃)'], 'o-')
ax1.set_title('2020년')
ax2.plot(temp_2021['월'],temp_2021['평균기온(℃)'], 'o-')
ax2.set_title('2021년')
ax3.plot(temp_2022['월'],temp_2022['평균기온(℃)'], 'o-')
ax3.set_title('2022년')
ax4.plot(temp_2023['월'],temp_2023['평균기온(℃)'], 'o-')
ax4.set_title('2023년')

plt.show()
```
