```python
data=data.dropna(subset=['addr'],how='any',axis=0)
```

```python
data['시']=data['addr'].str.split(" ").str.get(0)
```

```python
data_seoul = data[data['시']=='서울특별시']
data_seoul = data_seoul[['시','stat_nm']]
data_seoul.reset_index(inplace=True,drop=True)
```

```python
data_busan = data[data['시']=='부산광역시']
data_busan = data_busan[['시','stat_nm']]
data_busan.reset_index(inplace=True,drop=True)
```

```python
data_daegu = data[data['시']=='대구광역시']
data_daegu = data_daegu[['시','stat_nm']]
data_daegu.reset_index(inplace=True,drop=True)
```

```python
data_jeju = data[data['시']=='제주특별자치도']
data_jeju = data_jeju[['시','stat_nm']]
data_jeju.reset_index(inplace=True,drop=True)
```
