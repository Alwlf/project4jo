```python
import pandas as pd
import warnings
warnings.filterwarnings(action='ignore')
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 17
plt.rcParams['axes.unicode_minus']=False  
```

```python
data=pd.read_csv("행정구역_시군구_별_주민등록세대수_20230501153433.csv",encoding='cp949')
```

```python
data=data.rename(columns={'행정구역(시군구)별':'지역'})
data = data.loc[data['지역'].apply(lambda x: x in ['서울특별시','제주특별자치도']),['지역',"2020","2021"]]
data.reset_index(drop=True,inplace=True)
data=data.set_index('지역')
data=data.astype(int)
```

```python
data
```

```python
colors = ['lightskyblue', 'lightcoral']
plt.figure(figsize=(5,5))
plt.bar(data.index,data['2020'],color=colors)
plt.title('인구수(2020)')
plt.xlabel('지역')
plt.ylabel('인구수')
plt.ylim(0,6000000)
plt.show()
```

```python
colors =["lightgreen", "mediumseagreen"]
plt.figure(figsize=(5,5))
plt.bar(data.index,data['2021'],color=colors)
plt.title('인구수(2021)')
plt.xlabel('지역')
plt.ylabel('인구수')
plt.ylim(0,6000000)
plt.show()
```
