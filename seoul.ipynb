```python
seoul_type=seoul[['구','충전기구분','충전량']]
```

```python
# 급속 충전기
seoul_type_a = seoul_type[seoul_type['충전기구분']=='급속']
```

```python
seoul_type_a
```

```python
# 구별로 묶기
seoul_a = seoul_type_a.groupby('구')['충전량'].sum().reset_index()
```

```python
seoul_a
```

```python
# 완속 충전기
seoul_type_b = seoul_type[seoul_type['충전기구분']=='완속']
```

```python
seoul_b = seoul_type_b.groupby('구')['충전량'].sum().reset_index()
```

```python
seoul_b
```

```python
fig = plt.figure(figsize=(30,15))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.set_title("급속충전기")
ax2.set_title("완속충전기")
ax1.plot(seoul_a['구'],seoul_a['충전량'],c='r')
ax2.plot(seoul_b['구'],seoul_b['충전량'],c='b')
fig.suptitle("구별 종류별 충전기 개수")
fig.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(25,15))
x=seoul_a['구']
y1=seoul_a['충전량']
y2=seoul_b['충전량']
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 15
plt.rcParams['axes.unicode_minus']=False  
```

```python
fig, ax1 = plt.subplots(figsize=(20,5))
color_1 = 'tab:blue'
ax1.set_title('구별 충전기 종류', fontsize=16)
ax1.set_xlabel('구')
ax1.set_ylabel('급속충전기', fontsize=14, color=color_1)
ax1.plot(seoul_a['구'], seoul_a['충전량'], marker='s', color=color_1)
ax1.tick_params(axis='y', labelcolor=color_1)

ax2 = ax1.twinx()
color_2 = 'tab:red'
ax2.set_ylabel('완속충전기', fontsize=14, color=color_2)
ax2.plot(seoul_b['구'], seoul_b['충전량'], marker='o', color=color_2)
ax2.tick_params(axis='y', labelcolor=color_2)

fig.tight_layout()
plt.show()
```
