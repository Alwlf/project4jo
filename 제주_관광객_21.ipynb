```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

%matplotlib inline

rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
tl_1 = pd.read_excel('jejutour/2021년1월+관광객입도현황(확정치).xls')
tl_2 = pd.read_excel('jejutour/2021년+2월관광객입도현황(확정치).xls')
tl_3 = pd.read_excel('jejutour/2021년+3월관광객입도현황(확정치).xls')
tl_4 = pd.read_excel('jejutour/2021년+4월관광객입도현황(확정치).xls')
tl_5 = pd.read_excel('jejutour/2021년+5월관광객입도현황(확정치).xls')
tl_6 = pd.read_excel('jejutour/2021년+6월관광객입도현황(확정치).xls')
tl_7 = pd.read_excel('jejutour/2021년+7월관광객입도현황(확정치).xls')
tl_8 = pd.read_excel('jejutour/2021년+8월관광객입도현황(확정치).xls')
tl_9 = pd.read_excel('jejutour/2021년+9월관광객입도현황(확정치).xls')
tl_10 = pd.read_excel('jejutour/2021년+10월관광객입도현황(확정치).xls')
```

```python
# 제주도 2021년 내국인 관광객 각 월계 입도현황
tl_01 = tl_1.loc[4,["Unnamed: 6"]]
tl_02 = tl_2.loc[4,["Unnamed: 6"]]
tl_03 = tl_3.loc[4,["Unnamed: 6"]]
tl_04 = tl_4.loc[4,["Unnamed: 6"]]
tl_05 = tl_5.loc[4,["Unnamed: 6"]]
tl_06 = tl_6.loc[4,["Unnamed: 6"]]
tl_07 = tl_7.loc[4,["Unnamed: 6"]]
tl_08 = tl_8.loc[4,["Unnamed: 6"]]
tl_09 = tl_9.loc[4,["Unnamed: 6"]]
tl_10 = tl_10.loc[4,["Unnamed: 6"]]
```

```python
# 제주도 2021년 내국인 관광객 각 월계 입도현황 concat
tl_l_21 = pd.concat([tl_01,tl_02,tl_03,tl_04,tl_05,tl_06,tl_07,tl_08,tl_09,tl_10])
tl_l_21.to_frame()
```

```python
tl_l_211 = tl_l_21.to_frame()
```

```python
# 컬럼명 변환
tl_l_211.columns=['내국인']
tl_l_211
```

```python
# 인덱스명 변환
tl_l_211.index = ["1월","2월","3월","4월","5월","6월","7월","8월","9월","10월"]
```

```python
# 제주도 2021년 외국인 관광객 각 월계 입도현황
tl_01_f = tl_1.loc[12,["Unnamed: 6"]]
tl_02_f = tl_2.loc[12,["Unnamed: 6"]]
tl_03_f = tl_3.loc[12,["Unnamed: 6"]]
tl_04_f = tl_4.loc[12,["Unnamed: 6"]]
tl_05_f = tl_5.loc[12,["Unnamed: 6"]]
tl_06_f = tl_6.loc[12,["Unnamed: 6"]]
tl_07_f = tl_7.loc[12,["Unnamed: 6"]]
tl_08_f = tl_8.loc[12,["Unnamed: 6"]]
tl_09_f = tl_9.loc[12,["Unnamed: 6"]]
tl_10_f = tl_10.loc[12,["Unnamed: 6"]]
```
