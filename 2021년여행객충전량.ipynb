```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from matplotlib import font_manager, rc # 폰트 세팅을 위한 모듈 추가
font_path = "C:/Windows/Fonts/malgun.ttf" # 사용할 폰트명 경로 삽입
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)
```

```python
charge2021 = pd.read_csv('2021년제주충전량마지막.csv')
tour2021 = pd.read_excel('2021년제주여행객.xlsx')
```

```python
tour2021
```

```python
charge2021 = charge2021.drop('Unnamed: 0', axis=1)
tour2021 = tour2021.drop('Unnamed: 0', axis=1)
```

```python
charge2021
```

```python
tour2021.rename(columns = {'군구':'구'}, inplace=True)
tour2021.head(3)
```

```python
je = tour2021[tour2021['구'] == '제주시']
je[je['내/외국인'] == '합계'].reset_index()
```

```python
seo = tour2021[tour2021['구'] == '서귀포시']
seo[seo['내/외국인'] == '합계'].reset_index()
```

```python
tour2021 = tour2021.T
tour2021
```

```python
tour2021 = tour2021.drop(index='시도')
tour2021
```

```python
cols = list(range(2, 36, 3))  # 2부터 108까지 3씩 증가하는 인덱스 리스트
result = tour2021.iloc[:, cols]  # 해당 인덱스에 해당하는 열들만 선택
result
```

```python
jejudo2021 = result.T
jejudo2021
```

```python
jejudo2021[jejudo2021['구'] == '제주시']
```

```python
jeju2021 = jejudo2021.set_index('구')
jeju2021
```

```python
jeju2021 = jeju2021.groupby(['구']).sum().reset_index().replace({'구': {'서귀포시': '서귀포시'}})
jeju2021
```

```python
jeju2021 = jeju2021.drop(columns=['관광지', '내/외국인', '21년총계', '인원계'], axis=1)
```

```python
jeju2021.T
```

```python
new_jeju2021 = jeju2021.transpose()
new_jeju2021.columns = new_jeju2021.iloc[0]
new_jeju2021 = new_jeju2021.drop(new_jeju2021.index[0])
new_jeju2021
```

```python
# 제주도 구별 월별 총충전량
charge2021_monthly = charge2021.groupby(['구', 'month']).sum().reset_index()
charge2021_pivot = charge2021_monthly.pivot_table(values='충전량', index='month', columns='구', fill_value=0)
```

```python
charge2021_pivot = charge2021_pivot.rename(columns={'서귀포시': '서귀포시(총충전량)', '제주시': '제주시(총충전량)'})
charge2021_pivot
```

```python
# 6월평균값넣고 11월 12월
# 5월과 7월의 데이터를 추출하여 평균값 계산
mean_charging = charge2021_pivot.loc[['2021-05', '2021-07'], ['서귀포시(총충전량)', '제주시(총충전량)']].mean()

# 6월의 데이터에 평균값 채우기
charge2021_pivot.loc['2021-06', ['서귀포시(총충전량)', '제주시(총충전량)']] = mean_charging
```

```python
new_jeju2021 = new_jeju2021.rename(columns={'서귀포시': '서귀포시(여행객수)', '제주시': '제주시(여행객수)'})
```

```python
df = pd.concat([charge2021_pivot, new_jeju2021], axis=1)
df
```

```python
df = df.sort_index()
```

```python
df
```

```python
# 2021년 11월 12월 데이터삭제
df.drop(['2021-11','2021-12'], axis=0, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
obj_cols = df.select_dtypes(include='object').columns
df[obj_cols] = df[obj_cols].apply(pd.to_numeric, errors='coerce')
df = df.astype(float)
```

```python
df['제주도(총충전량)'] = df['서귀포시(총충전량)'] + df['제주시(총충전량)']
df['제주도(여행객수)'] = df['서귀포시(여행객수)'] + df['제주시(여행객수)']
```

```python
charge = df['제주도(총충전량)']
tourist = df['제주도(여행객수)']

fig = plt.figure(figsize=(14,7))
fig.suptitle('2021년 월별 총충전량 및 여행객수')

ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(charge.index, charge, label='제주도')
ax2.plot(tourist.index, tourist, label='제주도')

ax1.legend(loc='best')
ax2.legend(loc='best')

ax1.grid(True)
ax2.grid(True)

ax1.set_ylim(0, 500000)
ax2.set_ylim(0, 300000)

ax1.set_xlabel('월')
ax2.set_xlabel('월')

ax1.set_ylabel('총충전량')
ax2.set_ylabel('여행객수')

plt.show()
```
