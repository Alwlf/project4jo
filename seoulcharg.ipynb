```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc # 폰트 세팅을 위한 모듈 추가
font_path = "C:/Windows/Fonts/malgun.ttf" # 사용할 폰트명 경로 삽입
font = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font)
```

```python
df = pd.read_excel('서울전기차충전소.xlsx')
```

```python
df[['시', '구', '도로명주소']] = df['addr'].str.split(' ', n=2, expand=True)
df.head(3)
```

```python
seoul = df['구'].value_counts()
pd.DataFrame(seoul)
```

```python
print(df['구'].value_counts())
```

```python
chargeseoul = pd.value_counts(df['구'].values, sort=True, ascending=True)
chargeseoul.plot.bar(grid=True)
plt.title('서울충전기대수')
plt.xlabel('구')
plt.ylabel('충전기대수')
```
